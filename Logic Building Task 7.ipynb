1. Smart Light Controller

```python
class Smartlight:                 # Smart Light Controller
  def __init__(self, light_name):
    self.light_name = light_name
    self.is_on = False            # Default is OFF

  def turn_on(self):
    self.is_on = True

  def turn_off(self):
    self.is_on = False         # is_on stores light status

  def display_status(self):    # display the status of the light
    if self.is_on:
      print(f"{self.light_name} is ON")
    else:
      print(f"{self.light_name} is OFF")

# Object Creation
light = Smartlight("Bedroom Light")

# display_status: ON
light.turn_on()
light.display_status()

```

2. Employee ID Card System

```python
# Create a class named employee
class Employee:
  def __init__(self, name, emp_id, department):
    self.name = name
    self.emp_id = emp_id
    self.department = department

# Dispaly the details of emp_id
  def display_id_card(self):
   print("Employee id card details :")
   print("Name:", self.name)
   print("Emp_ID:", self.emp_id)
   print("Department:", self.department)

# Object creation
emp = Employee("Rahul", "EMP102", "IT" )
emp.display_id_card()

```

4. Product Price Tag Generator

```python
# Create the class named Price Tag
class Price_Tag:
  def __init__(self, product_name, price):
     self.product_name = product_name
     self.price = price

# Dispaly the price tag
  def display_Price_Tag(self):
    print("Product Details")
    print("Product Name:", self.product_name)
    print("Price:", "Rs", self.price)

# Object creation
price = Price_Tag("Headphones", 2499)
price.display_Price_Tag()

```

5. Movie Rating Display System
