1. Mobile recharge validation system

```python
# Function to validate_recharge

def validate_recharge(amount):
  valid_plans = [199, 299, 399, 599]

  if amount < 50:
    print("Recharge amount should be at least 50")
    return False
  elif amount not in valid_plans:
    print("Invalid recharge amount.")
    return False
  else:
    print("Recharge was successful")
    return True

# Retry using while loop
while True:
  amount = int(input("Enter recharge amount: "))
  if validate_recharge(amount):
    break       # exit loop if valid
  else:
    print("Please try again")

```

2. Inventary reorder alert system

```python
# Functions to check stock levels
def check_inventory(products):
  for item, stock in products.items():
    if stock < 15:
      print(item, ": Reorder alert")
    else:
      print(item, ": Stock OK")


# Simple dictionary
products = {"Rice":20, "Wheat":10, "Sugar":25, "Oil":12}

# fuction call
check_inventory(products)

```

3. Student result processing system

```python
# Fuction to process student result
def process_result(marks):
  total = 0

  # Loop to calculate total
  for mark in marks:
    total += mark

  average = total / len(marks)

  if average >= 50:
    return "Pass"

  else:
    return "Fail"

# example marks list
student_marks =  [60, 45, 70, 55]

# Function call
result = process_result(student_marks)

print("Average Marks:", sum(student_marks) / len(student_marks))
print("Result:", result)

```

4. Cab fair estimator with retry option

```python
def cab_fare(distance, peak):
  fare = 50 + (12 * distance)

  if peak == "yes":
    fare = fare * 1.25  # 25% extra

  return fare


# Input
distance = float(input("Enter distance:"))
peak = input("Peak hour? (yes/no):")

print("Total Fare: Rs", cab_fare(distance, peak))
```

5. Employee attendence eligibility checker

```python
def check_attendence(attendence):
  present = 0


  for day in attendence:
    if day == "P":
      present +=1

  percentage = (present / len(attendence)) * 100

  if percentage >= 75:
    return "Eligible"
  else:
    return "Not Eligible"

# Example
attendence_list = ["P", "A", "P", "P", "A", "P"]
print("Result:", check_attendence(attendence_list))
```

6. Password strength checker

```python
def check_password(password):
  has_digit = False
  has_special = False

  for ch in password:
    if ch.isdigit():
      has_digit = True
    if ch in "@#$":
      has_special = True

    if len(password) >=8 and has_digit and has_special:
      return "Strong Password"

    else:
      return "Weak Password"

 # Example
Pwd = input("Enter password:")
print(check_password(Pwd))
```
