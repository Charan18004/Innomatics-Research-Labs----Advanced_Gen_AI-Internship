```python
# marks list
marks = [48,75,90,33,60]
pass_count = 0
fail_count = 0

#using for we can check each mark and count how many are pass and fail

for m in marks:
  if m > 50:
    pass_count += 1
  else:
    fail_count += 1

print("Total_students :", len(marks)) # length of list
print("Pass_students :", pass_count) # no.of pass students
print("Fail_students :", fail_count) # no.of fail students


```

```python
names = [" Alice ", "bob", "CHARLIE"]

cleaned_names =[]

for name in names:
  cleaned = name.strip().lower()  #removes empty space and convert the list into lowercase
  cleaned_names.append(cleaned) #adds to the new list
  print(cleaned_names)
```

```python
messages = ["Hi", "Welcome to the Platform", "OK"] # list

for msg in messages: # takes one msg at a time
  length = len(msg)
print("Message :", msg )
print("Length :", length)

if length > 10 :   #msg greater than 10 falg it
  print("Long Message")
else:
  print() # blank space
```
