1. Social Media Post_Engagement Analyzer

```python
def analyze_engagement(likes_list):
  total_likes = 0

  # Using loop to calculate total likes
  for likes in likes_list:
    total_likes += likes

  print("Total Likes:", total_likes)

  if total_likes >= 1000:
    print("Post Status: Viral Post")
  else:
    print("Post Status: Normal Engagement")

  # Example input
  likes_data = [200, 300, 250, 300] # Total = 1050
  analyze_engagement(likes_data)
```

3. Agriculture- Rainfall Adequacy Checker

```python
def rainfall_checker(rainfall_data, required_level):  #  function to check whether rainfall adequate or not

  total = 0   # variable to store total rainfall

  # loop to calculate total rainfall
  for rain in rainfall_data:
    total += rain

  # Calculate average rainfall
  average = total / len(rainfall_data)

  # Print average rainfall
  print("Average Rainfall:", average)

  # Check if average meets required level
  if average >= required_level:
    print("Rainfall Status: Adequate Rainfall")

  else:
    print("Rainfall Status: Inadequate rainfall")

# Example data
rainfall = [70, 75, 80, 65, 70]

# Required minimum rainfall level
required = 70

# Call the function
rainfall_checker(rainfall, required)
```

6. Agriculture - Premium crop price filter

```python
# Function to filter premium crop prices
def premium_crop_filter(prices):

    premium_crops = []  # List to store premium crop prices

    # Loop through each price in the list
    for price in prices:

        # Check if price is greater than 2000
        if price > 2000:
            premium_crops.append(price)  # Add to premium list

    # Print premium crop prices
    print("Premium Crops:", premium_crops)


# Example crop prices
crop_prices = [1500, 2500, 1800, 3200, 1900]

# Call the function
premium_crop_filter(crop_prices)
```

7. System monitoring - Application health checker

```python
# Function to check application health based on error count
def check_application_health(errors):

    # Print the error count
    print("Error Count:", errors)

    # Check system health condition
    if errors == 0:
        print("System Status: Healthy")
    elif errors <= 5:
        print("System Status: Minor Issues")
    else:
        print("System Status: Critical Issues")


# Example error count
error_count = 7

# Call the function
check_application_health(error_count)
```

8. Banking - Daily transaction limit checker

```python
# Function to validate daily transaction amount
def validate_transaction(amount):

    daily_limit = 50000  # Set daily transaction limit

    # Print the transaction amount
    print("Transaction Amount:", amount)

    # Check if amount is within the daily limit
    if amount <= daily_limit:
        print("Transaction Status: Approved")
    else:
        print("Transaction Status: Rejected")


# Example transaction amount
transaction_amount = 60000

# Call the function
validate_transaction(transaction_amount)
```

9. E-learning -- Student attendence eligibility system

```python
# Function to calculate attendance percentage and check eligibility
def check_exam_eligibility(attendance_list):

    total_days = len(attendance_list)  # Total number of working days
    present_days = 0  # Counter for days present

    # Loop to count number of present days
    for day in attendance_list:
        if day == 1:   # 1 means Present, 0 means Absent
            present_days += 1

    # Calculate attendance percentage
    percentage = (present_days / total_days) * 100

    # Print attendance percentage
    print("Attendance Percentage:", percentage)

    # Check eligibility condition
    if percentage >= 75:
        print("Exam Eligibility: Eligible")
    else:
        print("Exam Eligibility: Not Eligible")


# Example attendance record (1 = Present, 0 = Absent)
attendance = [1, 1, 1, 0, 1, 1, 1, 1, 0, 1]  # 8 out of 10 present → 80%

# Call the function
check_exam_eligibility(attendance)
```

10. Smart Electricity bill analyzer

```python
# Function to calculate electricity bill and classify usage
def calculate_electricity_bill(units):

    bill = 0  # Variable to store total bill amount

    # Slab 1: First 100 units
    if units <= 100:
        bill = units * 3

    # Slab 2: 101–200 units
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)

    # Slab 3: Above 200 units
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)

    # Usage classification based on total bill
    if bill < 500:
        usage_status = "Low Usage"
    elif bill <= 1500:
        usage_status = "Moderate Usage"
    else:
        usage_status = "High Usage"

    return bill, usage_status  # Return both bill and status


# Example usage
units_consumed = 250

total_bill, status = calculate_electricity_bill(units_consumed)

print("Units Consumed:", units_consumed)
print("Total Bill:", total_bill)
print("Usage Level:", status)
```
