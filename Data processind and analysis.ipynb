```python
# 1. Employee Performance Bonus Eligibility

employees = {"Ravi":92,"Anitha":88, "Kiran":92, "Suresh":85}

# find thw highest score
highest_score = max(employees.values())

# find all employees with highest score
top_performers = []

for name, score in employees.items():
  if score == highest_score:
    top_performers.append(name)

# Display result
print("Top Performers Eligible for Bonus:",",".join(top_performers),"(Score:",highest_score, ")")
```

```python
# 2. Search Query Result Analysis

query = "Buy mobile phone buy phone online"

# Convert to lowercase and split words
words = query.lower().split()

# Count frequency
count = {}

for word in words:
  count[word] = count.get(word, 0) + 1

# Print words with frequency greater than 1
result = {}

for word in words:
  if count[word] > 1:
    result[word] = count[word]

print(result)
```

```python
# 3. Sensor Data Validation

sensor_readings = [3, 4, 7, 8, 10, 12, 5]

valid_readings = []

for index, value in enumerate(sensor_readings):
  if value % 2 == 0:    #check even number
      valid_readings.append((index, value))
print("Valid sensor readings (Hour, Value):")
print(valid_readings)
```

```python
# 4. Email domain usage analysis

emails = ["ravi@gmail.com", "anitha@yahoo.com", "kiran@gmail.com", "suresh@yahoo.com", "meena@yahoo.com"]

domain_count = {}

# Count domains
for email in emails:
  domain = email.split("@")[1]
  domain_count[domain] = domain_count.get(domain, 0) + 1

total = len(emails)

# Calculate and print percentage
for domain in domain_count:
  percentage = (domain_count[domain] / total) * 100
  print(domain + ":", str(int(percentage)) + "%")
```

```python
# 5. Sales spike detection

sales = [1200, 1500, 900, 2200, 1400, 3000]

# Calculate average
average = sum(sales) / len(sales)

# Check for sales 30% above average
for i in range(len(sales)):
  if sales[i] > average * 1.30:
    print("Day", i + 1, ":", sales[i])
```

```python
# 6. Duplicate user ID detection

user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

count = {}

# count the occurences in user_ids
for user in user_ids:
  count[user] = count.get(user, 0) + 1

# print duplicates only
for user in count:
  if count[user] > 1:
      print(user, "-->", count[user], "times" )
```
