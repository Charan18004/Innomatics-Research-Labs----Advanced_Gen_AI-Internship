1. Smart Parking Lot Management System

```python
def smart_parking_system(capacity, vehicle_logs):
    current_vehicles = 0
    peak_usage = 0

    for log in vehicle_logs:
        if log == "IN":
            current_vehicles += 1
        elif log == "OUT":
            if current_vehicles > 0:
                current_vehicles -= 1

        # Track peak usage
        if current_vehicles > peak_usage:
            peak_usage = current_vehicles

    print("Currently Parked Vehicles:", current_vehicles)
    print("Peak Parking Usage:", peak_usage)

    # Check capacity
    if current_vehicles > capacity:
        print("Parking Status: Capacity Exceeded ")
    else:
        print("Parking Status: Available ")


# Sample Input
capacity = 50
vehicle_logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]

smart_parking_system(capacity, vehicle_logs)

```

2. Online Food Delivery Time estimator

```python
# Function Definition
def delivery_time(distance, weather, traffic):

  # Base time calculation
  estimated_time = distance * 5

  # Traffic condition
  if traffic == "Low":
    estimated_time +=5
  elif traffic == "Medium":
    estimated_time +=10
  elif traffic == "High":
    estimated_time +=15

  # Weather condition
  if weather == "Normal":
    estimated_time +=0
  elif weather == "Rainy":
    estimated_time +=5
  elif weather == "Storm":
    estimated_time +=10

  print("Estimated Delivery Time :", estimated_time, "min")

  # Sample input
  distance = 8
  traffic = "High"
  weather = "Rainy"

  # Funtion call
  delivery_time(distance, weather, traffic)
```

3. Movie Theatre Seat Occupancy Analyzer

```python
# Function definition
def seat_occupancy_analyzer(total_seats, booked_seats):

  # Calculate no.of booked seats
  booked_count = len(booked_seats)

  # Calculate occupany percentage
  occupancy = (booked_count / total_seats) * 100

  print("Occupancy:", round(occupancy), "%")

  # Determine show status
  if occupancy == 100:
    print("show status : Housefull")
  elif occupancy >= 75:
    print("Show status : Almostfull")
    print("Suggestion : Consider opening additional shows")
  else:
    print("Show Status : Show is Available")

 # Sample input
total_seats = 200
booked_seats = [1] * 150

# function call
seat_occupancy_analyzer(total_seats, booked_seats)
```

4. Cloud Server Load Classification System

```python
# Function definition
def server_load_classification(cpu_readings):

   # Calculate average
   total = sum(cpu_readings)
   count = len(cpu_readings)
   average = total / count

   print("Average CPU reading :", round(average), "%")

   # Classify the rules
   if average < 50:
    print("Server Load : Normal")
   elif average > 50 and average < 80:
    print("Server Load : Warning")
   elif average > 80:
    print("Server Load : Critical")

    # Sample Input
    cpu_readings = [45, 60, 70, 85, 90]

    # Function call
    server_load_classification(cpu_readings)
```

6. Online Event registration Capacity Controller

```python
# Function to manage event registration
def manage_registration(capacity, registrations):

  # if registrations are within capacity
  if registrations <= capacity:
    confirmed = registrations
    waitlist = 0
    status = "Open"

  # If registrations exceed capacity
  else:
    confirmed = capacity
    waitlist = registrations - capacity
    status = "Closed"

  # Display results
  print("Confirmed Registrations:", confirmed)
  print("Waitlisted users:", waitlist)
  print("Registrations Status:", status)

  # sample input
  capacity = 100
  registrations = 105

  # Function call
  manage_registration(capacity, registrations)
```
